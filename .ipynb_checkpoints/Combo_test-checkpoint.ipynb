```python
import os
import numpy as np
import pandas as pd
```

```python
os.chdir('E:\Academics\Research\Data\Experimental data\Compression_test\Results_and_Analysis\Combo_test\Input')
Exp_Data = pd.read_excel('Experimental_Data_Iteration2.xlsx')
Exp_Data.head(5)
len(Exp_Data)
```

```python
os.chdir('E:\Academics\Research\Data\Experimental data\Compression_test\Results_and_Analysis\Combo_test\Output')

Combo_Modulus = pd.read_csv('Combo_Modulus_Iteration2.csv')
del Combo_Modulus['type']
del Combo_Modulus['rank']
Combo_Yield = pd.read_csv('Combo_Yield_Iteration2.csv')
del Combo_Yield['type']
del Combo_Yield['rank']
Combo_Ultimate = pd.read_csv('Combo_Ultimate_Iteration2.csv')
del Combo_Ultimate['type']
del Combo_Ultimate['rank']
Combo_Ductility = pd.read_csv('Combo_Ductility_Iteration2.csv')
del Combo_Ductility['type']
del Combo_Ductility['rank']

Combo_Data = Combo_Modulus.merge(Combo_Yield,on = ['id','Theta','Phi','Horizontal','Vertical'])
Combo_Data = Combo_Data.merge(Combo_Ultimate,on = ['id','Theta','Phi','Horizontal','Vertical'])
Combo_Data = Combo_Data.merge(Combo_Ductility,on = ['id','Theta','Phi','Horizontal','Vertical'])
print(Combo_Data.shape)

Combo_Data.head(5)
```

```python
#Filtering the projected values alone from Experimental + Projected values
Proj_Data = Combo_Data[Combo_Data['id'].isin(Exp_Data['id']) == False]
Proj_Data.head(5)
len(Proj_Data)
```

```python
# Computing Mean, Standard_deviation, and Mean +/- 2(standard deviation) for all the properties.
os.chdir('E:\Academics\Research\Data\Experimental data\Compression_test\Results_and_Analysis\Combo_test\Analysis')
properties = ['Youngs_Modulus','Yield_Strength','Ultimate_Strength','Ductility']

df = []

d = {}
e = {}

for i in range(0, len(properties)):
    Mean_Exp = np.mean(Exp_Data[properties[i]])
    stddev_Exp = np.std(Exp_Data[properties[i]], ddof= 1)

    Mean_Proj = np.mean(Proj_Data[properties[i]])
    stddev_Proj = np.std(Proj_Data[properties[i]])

    UL_2sigma_Exp = Mean_Exp + 2*stddev_Exp
    LL_2sigma_Exp = Mean_Exp - 2*stddev_Exp

    UL_2sigma_Proj = Mean_Proj + 2*stddev_Proj
    LL_2sigma_Proj = Mean_Proj - 2*stddev_Proj
    
    name = properties[i]+'_Outliers'
    
    e[name] = Exp_Data[(Exp_Data[properties[i]] > UL_2sigma_Exp) | (Exp_Data[properties[i]] < LL_2sigma_Exp)]
    
    d[name] = Proj_Data[(Proj_Data[properties[i]] > UL_2sigma_Proj) | (Proj_Data[properties[i]] < LL_2sigma_Proj)]
    
    Outliers_Exp_Count = len(e[name])
    Outliers_Proj_Count = len(d[name])
    
    df.append((properties[i],Mean_Exp,Mean_Proj,stddev_Exp,stddev_Proj,
               (UL_2sigma_Exp,LL_2sigma_Exp),(UL_2sigma_Proj,LL_2sigma_Proj),Outliers_Exp_Count,Outliers_Proj_Count ))
              
statistics = pd.DataFrame(df,columns = ('Property','Mean_Exp','Mean_Proj','Std_dev_Exp',
                                  'Std_dev_Proj','Mean+/-2sigma_Exp','Mean+/-2sigma_Proj','Outliers_Exp','Outliers_Proj'))
statistics
              
# statistics.to_csv('Statistics_Iteration1.csv')
```

# Finding the occurences of outliers on all four properties

```python
#Seperating Outliers corresponding to each property
Youngs_Modulus_Outliers = d['Youngs_Modulus_Outliers']
Yield_Strength_Outliers = d['Yield_Strength_Outliers']
Ultimate_Strength_Outliers = d['Ultimate_Strength_Outliers']
Ductility_Outliers = d['Ductility_Outliers']
```

```python
# Merge all outliers to find the unique ids accross all four different properties.
Master_Outliers = Youngs_Modulus_Outliers.merge(Yield_Strength_Outliers, on = ['id','Theta','Phi','Horizontal','Vertical','Youngs_Modulus','Yield_Strength','Ultimate_Strength','Ductility'],how = 'outer')
Master_Outliers = Master_Outliers.merge(Ultimate_Strength_Outliers, on = ['id','Theta','Phi','Horizontal','Vertical','Youngs_Modulus','Yield_Strength','Ultimate_Strength','Ductility'], how = 'outer')
Master_Outliers = Master_Outliers.merge(Ductility_Outliers, on = ['id','Theta','Phi','Horizontal','Vertical','Youngs_Modulus','Yield_Strength','Ultimate_Strength','Ductility'], how = 'outer')
```

```python
Master_id = np.array(Master_Outliers['id'])
Modulus_id = np.array(Youngs_Modulus_Outliers['id'])
Yield_id = np.array(Yield_Strength_Outliers['id'])
Ultimate_id = np.array(Ultimate_Strength_Outliers['id'])
Ductility_id = np.array(Ductility_Outliers['id'])

sample_id = []
count = []
Modulus = []
Yield = []
Ultimate = []
Ductility = []

for i in range(0, len(Master_id)):
    k = 0
    Modulus.append(0)
    Yield.append(0)
    Ultimate.append(0)
    Ductility.append(0)
    for j in range(0,len(Modulus_id)):
        if int(Master_id[i]) == int(Modulus_id[j]):
            k = k+1
            Modulus[i] = 1
    for j in range(0,len(Yield_id)):
        if int(Master_id[i]) == int(Yield_id[j]):
            k = k+1
            Yield[i] = 1
    for j in range(0,len(Ultimate_id)):
        if int(Master_id[i]) == int(Ultimate_id[j]):
            k = k+1
            Ultimate[i] = 1
    for j in range(0,len(Ductility_id)):
        if int(Master_id[i]) == int(Ductility_id[j]):
            k = k+1
            Ductility[i] = 1
    sample_id.append(int(Master_id[i]))
    count.append(k)

df = {'sample_id':sample_id,'count':count,'Modulus':Modulus,'Yield':Yield,'Ultimate':Ultimate,'Ductility':Ductility}

Outlier_Occurence = pd.DataFrame(df,columns = ['sample_id','count','Modulus','Yield', 'Ultimate','Ductility'])

Outlier_Occurence_1 = Outlier_Occurence[Outlier_Occurence['count'] ==1]
Outlier_Occurence_2 = Outlier_Occurence[Outlier_Occurence['count'] ==2]
Outlier_Occurence_3 = Outlier_Occurence[Outlier_Occurence['count'] ==3]
Outlier_Occurence_4 = Outlier_Occurence[Outlier_Occurence['count'] ==4]

df1 =  {'One_Occerence':[len(Outlier_Occurence_1)],'Two_Occuerence':[len(Outlier_Occurence_2)],
        'Three_Occurence':[len(Outlier_Occurence_3)],'Four_Occurence':[len(Outlier_Occurence_4)]}

Outlier_Frequency = pd.DataFrame(df1,columns = ['One_Occerence','Two_Occuerence','Three_Occurence','Four_Occurence'])
Outlier_Frequency
# Outlier_Frequency.to_csv('Outlier_Frequency.csv')
```

# Finding the distribution of outliers accross locations of plate

240 x 240 mm plate is divided into four quandrants of 120 x 120 each. And Number of samples in each quadrants are counted. 

```python

orientations = pd.DataFrame({'Theta': [0,45,90,0,90,270,180,0,135],'Phi': [90,90,90,0,45,45,45,45,90]},
                            columns =['Theta','Phi'])

# def Outliers_Distribution4(Outliers_Data):

#     df = []
#     for i in range(0,9):
#         orientation = Outliers_Data[(Outliers_Data.Phi == orientations['Phi'][i]) & 
#                                           (Outliers_Data.Theta == orientations['Theta'][i])]
#         Phi = orientations['Phi'][i]
#         Theta = orientations['Theta'][i]
#         orientation_TL = orientation[(orientation.Vertical > 120) & (orientation.Horizontal > 120)]
#         orientation_TR = orientation[(orientation.Vertical > 120) & (orientation.Horizontal < 120)]
#         orientation_BL = orientation[(orientation.Vertical < 120) & (orientation.Horizontal > 120)]
#         orientation_BR = orientation[(orientation.Vertical < 120) & (orientation.Horizontal < 120)]
#         df.append((Theta,Phi,len(orientation), len(orientation_TL),len(orientation_TR),len(orientation_BL),len(orientation_BR)))

#     Distribution = pd.DataFrame(df,columns = ('Theta','Phi','Total','Top_Left','Top_Right','Bottom_Left','Bottom_Right'))

#     return(Distribution)

Occurence4_id = np.array(Outlier_Occurence_4['sample_id'])
Occurence3_id = np.array(Outlier_Occurence_3['sample_id'])
Occurence2_id = np.array(Outlier_Occurence_2['sample_id'])
Occurence1_id = np.array(Outlier_Occurence_1['sample_id'])

#Compares two different list of numbers and returns the common numbers

def id_compare(orient,Occurence_id):
    pref_id = []
    Dataframe_id = np.array(orient['id'])
    for i in range(0,len(Dataframe_id)):
        for j in range(0,len(Occurence_id)):
            if Dataframe_id[i] == Occurence_id[j]:
                pref_id.append(Dataframe_id[i])
    return(pref_id)

# Returns the sample ids in the order of preference. 
# If the sample to be tested is in the 3 times occurence it returns that samples id
# If the sample to be tested is not in 3 times occurence it searches in 2 times occurence
# if it is there it ruturns those ids
# Least preference is for one time occurence. 
# If the sample id is not available in 3 time or 2 time occurences it returns the ids in 1 time occurence

def preferred_id(orient):
    pref_id = id_compare(orient,Occurence4_id)
    if len(pref_id) > 0:
        return(pref_id)
    else:
        pref_id  = id_compare(orient,Occurence3_id)
        if len(pref_id) >0:
            return(pref_id)
        else:
            pref_id  = id_compare(orient,Occurence2_id)
            if len(pref_id) >0:
                return(pref_id)
            else:
                pref_id = id_compare(orient,Occurence1_id)
                return(pref_id)
    return(pref_id)

# Returns the sample ids in the order of preference. 
# If the sample to be tested is in the 3 times occurence it returns that samples id
# If the sample to be tested is not in 3 times occurence it searches in 2 times occurence
# if it is there it ruturns those ids
# Least preference is for one time occurence. Finaly it returns the sample id in one time occurence.

def Priority_Distribution(Outliers_Data):

    df = []
    df1 = []
    for i in range(0,9):
        orientation = Outliers_Data[(Outliers_Data.Phi == orientations['Phi'][i]) & 
                                          (Outliers_Data.Theta == orientations['Theta'][i])]
        Phi = orientations['Phi'][i]
        Theta = orientations['Theta'][i]
        orientation_TL = orientation[(orientation.Vertical > 120) & (orientation.Horizontal > 160)]
        TL = preferred_id(orientation_TL)
        TL1 = len(orientation_TL)
        orientation_TC = orientation[(orientation.Vertical > 120) & (orientation.Horizontal > 80) & (orientation.Horizontal < 160)]
        TC = preferred_id(orientation_TC)
        TC1 = len(orientation_TC)
        orientation_TR = orientation[(orientation.Vertical > 120) & (orientation.Horizontal > 0) & (orientation.Horizontal < 80)]
        TR = preferred_id(orientation_TR)
        TR1 = len(orientation_TR)
        orientation_BL = orientation[(orientation.Vertical < 120) & (orientation.Horizontal > 160)]
        BL = preferred_id(orientation_BL)
        BL1 = len(orientation_BL)
        orientation_BC = orientation[(orientation.Vertical < 120) & (orientation.Horizontal > 80) & (orientation.Horizontal < 160)]
        BC = preferred_id(orientation_BC)
        BC1 = len(orientation_BC)
        orientation_BR = orientation[(orientation.Vertical < 120) & (orientation.Horizontal > 0) & (orientation.Horizontal < 80)]
        BR = preferred_id(orientation_BR)
        BR1 = len(orientation_BR)
        df.append((Theta,Phi,len(orientation), TL,TC,TR,BL,BC,BR))
        df1.append((Theta,Phi,len(orientation),TL1,TC1,TR1,BL1,BC1,BR1))

    Priority = pd.DataFrame(df,columns = ('Theta','Phi','Total','Top_Left','Top_Center','Top_Right','Bottom_Left','Bottom_Center','Bottom_Right'))
    Distribution = pd.DataFrame(df1,columns = ('Theta','Phi','Total','Top_Left','Top_center','Top_Right','Bottom_Left','Bottom_center','Bottom_Right'))
    return(Priority,Distribution)
```

```python
#Converts sample_id(1 to 624) to the alphanumeric(A1 to Y25) name on plate
Locations = ['Top_Left','Top_Center','Top_Right','Bottom_Left','Bottom_Center','Bottom_Right']

Num_to_alphabet = pd.read_csv('E:\Academics\Research\Data\Experimental data\Compression_test\Results_and_Analysis\Combo_test\Input\Num_to_alphabet.csv')


def numtoalpha(Outlier_Priority):
    for k in range(0, len(Outlier_Priority)):
        for i in (Locations):
            if len(Outlier_Priority[i][k])> 0:
                for j in range(0,len(Outlier_Priority[i][k])):
                    Outlier_Priority[i][k][j] = Num_to_alphabet['Sample'][int(Outlier_Priority[i][k][j])]
            else:
                Outlier_Priority[i][k] =[]
    return(Outlier_Priority)
                
```
