```python
#Modulus
mu_exp = 180
mu_proj = 142
sig_exp = 78
sig_proj = 82

mu_exp2 = 205
mu_proj2 = 216
sig_exp2 = 74
sig_proj2 = 92

x = np.linspace(mu_exp2-2*sig_exp,mu_exp2+2*sig_exp, 100)
x1 = np.linspace(mu_proj2-2*sig_proj,mu_proj2+2*sig_proj, 100)

y = np.linspace(mu_exp2-2*sig_exp2,mu_exp2+2*sig_exp2, 100)
y1 = np.linspace(mu_proj2-2*sig_proj2,mu_proj2+2*sig_proj2, 100)

fig = plt.figure(figsize = (15,5))

fig.subplots_adjust(wspace=.5)

ax1 = fig.add_subplot(121)
ax1.plot(x,mlab.normpdf(x, mu_exp, sig_exp),'r-',label = 'Experiment')
ax1.plot(x1,mlab.normpdf(x1, mu_proj, sig_proj),'b-', label = 'Model')
ax1.set_xlabel('Modulus (Gpa)', size = 15, labelpad = 10)
ax1.set_ylabel('pdf', size = 15)
ax1.set_title('Iteration1')
plt.legend()

ax2 = fig.add_subplot(122,sharex = ax1,sharey = ax1)
ax2.plot(y,mlab.normpdf(y, mu_exp2, sig_exp2),'r-',label = 'Experiment')
ax2.plot(y1,mlab.normpdf(y, mu_proj2, sig_proj2),'b-',label = 'Model')
ax2.set_xlabel('Modulus (Gpa)', size = 15, labelpad = 10)
ax2.set_ylabel('pdf', size = 15)
ax2.set_title('Iteration2')
plt.legend()



plt.savefig('C:\Users\senth\Desktop\Images\modulus.jpg')
```
