```python
import os
import numpy as np
import pandas as pd
import shutil
import matplotlib.pyplot as plt
from numpy import arange
import matplotlib.patches as mpatches
```

```python
os.chdir('E:\Academics\Research\Data\Experimental data\Tensile_test\Results')
Data1 = pd.read_excel('Block2.xlsx', sheet = 1).dropna()
Data2 = pd.read_excel('Block3.xlsx', sheet = 1).dropna()
```

```python
frames = [Data1,Data2]
Data = pd.concat(frames)
Data
```

```python
# map_orientation = {'ZXY': 1, 'ZYX': 2,'XZY': 3,'YZ':4}
# map_block = {2:'g',3:'b'}

# tick_lab = ['ZXY', 'ZYX','YZX','XZY']
# tick_pos = range(1,5,1)
# tick_pos

map_orientation = {'ZXY': 'g', 'ZYX': 'b','XZY': 'r','YZX':'c'}
map_block = {2:'o' , 3: '^'}
```

```python
def add_legend(label, **kwds):
    return mpatches.Patch(label=label, **kwds)
```

```python
fig, ax = plt.subplots(figsize = (20,15),dpi = 160)
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize =20)

for row in Data.itertuples():
    ax.plot(row.From_Edge, row.Ductility,
              marker= map_block[row.Block],color = map_orientation[row.Orientation], markersize = 10)
    
ZXY = add_legend('ZXY',color = 'g')   
ZYX = add_legend('ZYX',color = 'b')
XYZ = add_legend('XZY',color = 'r')
YZX = add_legend('YZX',color = 'c')
    
        
ax.set_xlabel('Distance from Edge (mm)',fontsize = 25,labelpad = 5)
ax.set_ylabel('Ductility(%)',fontsize =25,labelpad = 10)

# ax.set_xticks(tick_pos)
# ax.set_xticklabels(tick_lab)
ax.set_xlim([-10,60])
plt.legend(handles=[ZXY, ZYX, XYZ, YZX],
           bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)

plt.show()
```
